### sharing some subject knowledge it is clear that , a attacker will have certain motive , the attacker will stay at the server for some time only , so the function was really helpfull in checking this thing , we will now use a simplified way for using the aggreate , with a variable count and an accumultor , we will use 2 accumulators for that , one will do a sum as done above and one will keep the normal /attack count , that we had to do in a separate window to find the count and find the mean , then we will divide the both to find the mean 

```python
#n3=rdd.map(lambda x:x.split(','))
#n1=n3.filter(lambda x:x[41]=='normal.')
#a2=n3.filter(lambda x:x[41]!='normal.')
#a2.count()
#norm3=n2.map(lambda x:int(x[0]))
#norm3.collect()
#attk2=a2.map(lambda x:int(x[0]))

# this part will be there , now we have 2 rdd's we will work the agreegate now 

norm_agg=norm3.aggregate((0,0),(lambda acc,value:(acc[0]+value,acc[1]+1)),(lambda acc1,acc2:(acc1[0]+acc2[0],acc1[1]+acc2[1])))
print('the normal mean duration was :',norm_agg[0]/norm_agg[1])


```

# Key _value pair RDD's -Dictionary

### above in the sheet we did a parse interaction , in that we used a tag for the rdd ,by putting the value of a feature in the front , that is what we will do here using key value pair

```python
kv=rdd.map(lambda x:x.split(','))
kvdi=kv.map(lambda x:(x[41],x))
kvdi.take(1)
```

#### above we used a reduce , that helped in adding the values , and then we created a separate function 2 create 2 accumulators and then add the values in one of them , now we will do the same using the key value pair accumulators by using reducebykey

```python
kv_agg=kv.map(lambda x:(x[41],float(x[0])))
kv_agg_dur=kv_agg.reduceByKey(lambda x,y:x+y)### this will provide the duration for each network interaction , simply the 1st accumalator that adds the value
kv_agg_dur.collect()
```

### we can see that the normal above is 21075991.0 , that is the 1st accumulator work , we also had a 2nd accumulator that is nothing but gives the count for the interaction , above we did for just normal , now we do for whole , using key value pair

```python
#kv=rdd.map(lambda x:x.split(','))
#kvdi=kv.map(lambda x:(x[41],x)) this will remain as now the normal , attack enteries work as a key , clear as we did above as well
cou_key=kvdi.countByKey()
```

```python
cou_key
```

## it is cleat that reduce by key  works on rows , and the counts by key  works on columns

## now we know what to do , we have both the values we can find the mean durations, although i thought on it that we have rdd's and how are we going to divide each element corresponding to the exact entry to get the mean , then it was clear by combine by key , that will keep the values as a key and a value in a pair for each element then collect as map , shows the pair , now it will be easy to do any aggregation be it div , mul or mean  

```python
com_key=kv_agg.combineByKey((lambda x:(x,1)),
                                (lambda acc,value:(acc[0]+value,acc[1]+1)),
                                (lambda acc1, acc2: (acc1[0]+acc2[0], acc1[1]+acc2[1])))
com_key.collectAsMap()
```

### we have the key and the values combined now , we will use the key and the value to calculate the mean for all the network int types 

```python
dur_mean_in=com_key.map(lambda k :(k[0],k[1][0]/k[1][1])).collectAsMap()
dur_mean_in
```

### printing as sorted values

```python
for tag in sorted(dur_mean_in,key=dur_mean_in.get,reverse=True):
  print(tag,dur_mean_in[tag])
```

#### as we have text field in the data , we will change that into the numerical by separating it from the RDD , and then use the map function to reverse back the numerical values

```python
csv_data = rdd.map(lambda x: x.split(","))
test_csv_data = test_raw_data.map(lambda x: x.split(","))

protocols = csv_data.map(lambda x: x[1]).distinct().collect()
services = csv_data.map(lambda x: x[2]).distinct().collect()
flags = csv_data.map(lambda x: x[3]).distinct().collect()
```

```python
def create_labeled_point(line_split):
    # leave_out = [41]
    clean_line_split = line_split[0:41]
    
    # convert protocol to numeric categorical variable
    try: 
        clean_line_split[1] = protocols.index(clean_line_split[1])
    except:
        clean_line_split[1] = len(protocols)
        
    # convert service to numeric categorical variable
    try:
        clean_line_split[2] = services.index(clean_line_split[2])
    except:
        clean_line_split[2] = len(services)
    
    # convert flag to numeric categorical variable
    try:
        clean_line_split[3] = flags.index(clean_line_split[3])
    except:
        clean_line_split[3] = len(flags)
    
    # convert label to binary label
    attack = 1.0
    if line_split[41]=='normal.':
        attack = 0.0
        
    return lp(attack, np.array([float(x) for x in clean_line_split]))

training_data = csv_data.map(create_labeled_point)
test_data = test_csv_data.map(create_labeled_point)
```

```python
training_data.take(5)
```

```python
test_data.take(5)
```

```python
from pyspark.mllib.tree import DecisionTree, DecisionTreeModel
from time import time

# Build the model
t0 = time()
tree_model = DecisionTree.trainClassifier(training_data, numClasses=2, 
                                          categoricalFeaturesInfo={1: len(protocols), 2: len(services), 3: len(flags)},
                                          impurity='entropy', maxDepth=4, maxBins=100)
tt = time() - t0

print ("Classifier trained in {} seconds".format(round(tt,3)))
```

```python
predictions = tree_model.predict(test_data.map(lambda p: p.features))
labels_and_preds3 = test_data.map(lambda p: p.label).zip(predictions)
```

```python
t0 = time()
test_accuracy = labels_and_preds3.filter(lambda x:x[0]==x[1]).count() / float(test_data.count())
tt = time() - t0

print ("Prediction made in {} seconds. Test accuracy is {}".format(round(tt,3), round(test_accuracy,4)))
```

### we have a good time and good accuracy than the log regg
